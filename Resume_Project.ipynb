2. Resume: It contains the actual text or content of the resumes. It includes the details of the candidates' education, work experience, skills, achievements, and any other relevant information typically found in a resume.

The primary goal of the project is to automate the categorization of candidates into specific fields such as Engineer, Data Scientist, Java Developer or other relevant categories based on the content of their resumes.

### Step 4: Standardizing Data:

Perform data standardization and cleaning operations based on specific requirements. This can include tasks like converting units, formatting dates or strings, removing special characters, or dealing with inconsistent data.

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

### Step 6: Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df.Category.unique()
```

### Step 7: Text Vectorization with TF-ID

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requredTaxt  = tfidf.transform(df['Resume'])
```

## ML Model Building

Stages:

* Step1: Train and Test data split
* Step2: Training the model
* Step3: Testing the model
* Step4: Evaluating Model Performance

### Step1: Train and Test data split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requredTaxt, df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Step2: Training the model

```python
# importing the algorithm

from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Fit the model on the training data

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
```

### Step3: Testing the model

```python
#Predicting the target: Category against the predictor in the training data set
#Predicted data stored in y_pred_train

y_pred_train = clf.predict(X_train)
```

```python
#Predicting the target: Category against the predictor in the testing data set
#Predicted data stored in y_pred_test

y_pred_test = clf.predict(X_test)
```

```python
predicted = clf.predict(X_test)
```
